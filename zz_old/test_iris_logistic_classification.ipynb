```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from logistic_classification import LogisticClassification
import numpy as np
import matplotlib.pyplot as plt
from learningcurve_plots import learningcurve_plots
```

```python
iris = load_iris()
X, y = iris.data, iris.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
model_setosa = LogisticClassification(lr=0.01, n_iters=2000)
```

```python
# setosa vs not setosa
y_train_setosa = np.where(y_train == 0, 1, 0)
y_test_setosa = np.where(y_test == 0, 1, 0)
```

```python
model_setosa.fit(X_train, y_train_setosa)
```

```python
learningcurve_plots(model_setosa.cost_hist, logy=True)
```

```python
y_pred_setosa = model_setosa.predict(X_test)
accuracy_score(y_test_setosa, y_pred_setosa)
```
